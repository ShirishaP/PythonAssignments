Create a data frame and merge it with new column

```python
import pandas as pd
name = ["Indu","Siva","Kiran","Deepak","chowman","Rob","Bob","Kim","Satya","Hillman"]
age = [20,25,23,45,22,31,43,56,65,89]
idNum = [1001,1002,1003,1004,1005,1006,1007,1008,1009,1010]
dict1 = {
    'Name' : name,
    "age" : age,
    "idNum" : idNum
}

df = pd.DataFrame(dict1)
df
city = ['Banglore','Pune','Hyderabad','Mumbai',"Mysore","Trivendrum","Chennai","Agra","Delhi","Chandigarh"]

df["City"] = city
df
```

```python
name = ['Suv','Mounica','Joey']
room_number = [101,102,103]
dict2={
    "Name":name,
    "RoomNo":room_number
}
df2 = pd.DataFrame(dict2)
df2
```

```python
pd.merge(df,df2,on='Name',how='outer')
```
